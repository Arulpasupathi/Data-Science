```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def plot_curve(func):
    x=np.linspace(-10,10,1000)
    y=[func(i) for i in x]
    plt.plot(x,y)
    plt.title(func.__name__)
```

```python
func=lambda x: x**2
```

```python
plot_curve(func)
```

```python
def plot_another_curve(funs):
    x=np.linspace(-10,10,1000)
    delta=0.01
    y=(funs(x+delta)-funs(x))/delta
    plt.plot(x,y)
```

```python
funcs=lambda x: x**2
plot_another_curve(funcs)
```
